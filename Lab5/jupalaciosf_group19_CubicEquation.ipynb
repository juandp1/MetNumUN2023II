# Explanation
La función solve_cubic resuelve una ecuación cúbica de la forma $x^3 + bx^2 + cx + d = 0$
Los coeficientes $b,c$ y $d$ se pasan como argumentos a la función. La función devuelve las raíces de la ecuación cúbica en forma de tupla ($x_1$, $x_2$, $x_3$).

Pasos de la función `solve_cubic`:

Calcula varios valores intermedios, como p, q, y Δ, que se utilizan para determinar el tipo de raíces de la ecuación cúbica.

La función verifica el valor de Δ:

> Si Δ > 0, significa que hay una raíz real y dos raíces complejas. Calcula las raíces reales y las raíces complejas correspondientes utilizando fórmulas específicas. Las raíces complejas se almacenan en formato complex.

> Si Δ = 0, significa que hay tres raíces reales, y dos de ellas son iguales. Calcula las tres raíces utilizando fórmulas específicas.

>Si Δ < 0, significa que hay tres raíces reales distintas. Utiliza la fórmula de De Moivre para calcular las tres raíces.

Después de calcular las raíces, se la función `cubic` para verificar si las raíces son válidas.

Finalmente, las raíces ajustadas se devuelven como una tupla ($x_1$, $x_2$, $x_3$).
